```python
import ktrain
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import classification_report

MAX_LEN = 10000
CATEGORIES_LIST = ['Arte e Design', 'Música e Áudio', 'Comunicação', 'Finanças', 'Entretenimento', 'Medicina', 'Comer e Beber', 'Reproduzir e Editar Videos', 'Mapas e Navegação', 'Social', 'Fotografia', 'Compras', 'Produtividade', 'Educação', 'Livros e Referência', 'Encontro', 'Turismo e Local', 'Ferramentas']
NAME_TERMINAL = "ep2"
print("start")
```

```python
#Importando a biblioteca pandas e carregando as amostras
df_test = pd.read_pickle('./Amostras_Teste.pkl')
print(df_test.shape)
df_test = df_test.dropna()
print(df_test.shape)
print("importou as bibliotecas")
```

```python
#Load predictor
p = ktrain.load_predictor('./bert-base-multilingual-cased_e2_apps.model')
print("carregou o modelo")
```

```python
save_classification_context = []
save_classification_non_context = []
for this_category in CATEGORIES_LIST:
    df_aux = df_test[df_test['category']==this_category]
    if len(df_aux) > MAX_LEN:
        df_aux = df_aux.head(MAX_LEN)
    x_test_context = df_aux[['review', 'category']].values
    x_test_context = list(map(tuple, x_test_context))
    x_test_non_context = df_aux.review.values
    y_true = df_aux.label_feelings.values
    print(NAME_TERMINAL, this_category)
    #Predições com contexto
    y_pred_context = p.predict(x_test_context)
    print("Métricas com contexto:")
    save_classification_context.append(classification_report(y_true[0:len(y_pred_context)], y_pred_context))
    print(save_classification_context[-1])
    #Predições sem contexto
    y_pred_non_context = p.predict(x_test_non_context)
    print("Métricas sem contexto:")
    save_classification_non_context.append(classification_report(y_true[0:len(y_pred_non_context)], y_pred_non_context))
    print(save_classification_non_context[-1])
    
    print("Terminando processo de:", NAME_TERMINAL, this_category)
    print()
    
print("Encerrada a execução ----------------------------------------------------------")
```

```python
print(NAME_TERMINAL)
for i in range(len(save_classification_context)):
    print('context -', CATEGORIES_LIST[i])
    print(save_classification_context[i])
    print('non context -', CATEGORIES_LIST[i])
    print(save_classification_non_context[i])
```
